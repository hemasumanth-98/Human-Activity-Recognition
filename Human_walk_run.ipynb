```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image  import ImageDataGenerator
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import softmax
from keras.api._v2.keras import activations
import numpy as np
```

```python
train_walk_run = ImageDataGenerator(rescale=1./255, shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
train_data = train_walk_run.flow_from_directory(
'/content/train',
target_size=(224,224),
batch_size=15,
class_mode='categorical')
```

```python
test_data = train_walk_run.flow_from_directory(
'/content/test',
target_size=(224, 224),
batch_size=15,
class_mode='categorical')
```

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.layers import Flatten
```

```python
Image_size=[224,224]
```

```python
sol=VGG16(input_shape=Image_size + [3] , weights='imagenet' , include_top = False)
```

```python
for i in sol.layers:
  i.trainable = False
```

```python
y=Flatten()(sol.output)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import softmax
```

```python
from keras.api._v2.keras import activations
final = Dense(2, activation = 'softmax')(y)
```

```python
from tensorflow.keras.models import Model
```

```python
vgg16_model = Model(sol.inputs, final)
```

```python
vgg16_model.summary()
```

```python
vgg16_model.compile(optimizer = 'adam' , loss = 'categorical_crossentropy' , metrics = ['Accuracy'])
```

```python
vgg16_model.fit(train_data, epochs = 5,validation_data=test_data)
```

```python
vgg16_model.save('CNN .h5')
```
